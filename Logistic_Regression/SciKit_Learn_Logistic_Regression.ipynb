```python
#Logistic Using Scikit

import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
#data for classification

X,y = make_blobs( centers = [[1,1], [-1,-1]], cluster_std = 1.5, random_state = 0 )
```

```python
#visualizing

plt.figure(figsize=[14,6])
plt.scatter( X[:,0], X[:,1], c=y, cmap= plt.cm.bwr)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
```

```python
#using scikit 

from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(solver = 'lbfgs')
clf.fit(X,y)

y_pred = clf.predict(X)

acc = metrics.accuracy_score(y, y_pred)

print(f'Accuracy of model given = {acc}')
```

```python
#visualising


plt.figure(figsize=[14,6])

x1, x2 = np.meshgrid(np.linspace(-5,5), np.linspace(-5,5))

y_prob = clf.predict_proba(np.hstack((x1.reshape(-1,1), x2.reshape(-1,1))))[:,1]

plt.contourf(x1, x2, y_prob.reshape(50,50), cmap = plt.cm.bwr, alpha = 0.75, vmin = 0, vmax = 0.95)

plt.colorbar()

plt.scatter(X[:,0], X[:,1], c=y, edgecolors = 'white', cmap = plt.cm.bwr)
```
